# Train - Test Validation Mechanism

```python
img = mpimg.imread('TrainTestSpilit.png')
plt.figure(figsize=(15,15))
plt.imshow(img)
plt.show()
```

```python
predictors = ['Age_08_04', 'KM', 'Fuel_Type', 'HP', 'Quarterly_Tax', 'Weight']
target = 'Price'

Data_X = pd.get_dummies(car_df[predictors], drop_first=True).drop(columns = ['Fuel_Type_Diesel'])
Data_y = car_df.Price

X_train, X_test, y_train, y_test = train_test_split(Data_X,Data_y,  test_size=0.3, random_state=1)
```

```python
# Data Before
print(Data_X.shape)
print(Data_y.shape)
```

```python
# Data After
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Design of Experiments

We will split the dataset into a trainset (%70) and a testset(%30), and predict the price of the testset with the following three methods and compare their performance:

    - M1 Radnom between min and max
    - M2 Naive mean
    - M3 LR

```python
result_df = pd.DataFrame(index=X_test.index, columns=['M1','M2','M3','Actual'])
result_df
```

```python
result_df.Actual = y_test
result_df
```

### Predict Using M2 Naive Mean

```python
result_df.M2 = y_train.mean()
result_df
```

### Predict Using M3 Random

```python
result_df.M1 = np.random.uniform(y_train.min(), y_train.max(),y_test.shape)
result_df
```

```python
result_df['|M1-Actual|'] = abs(result_df.M1-result_df.Actual)
result_df['|M2-Actual|'] = abs(result_df.M2-result_df.Actual)
result_df
```

```python
plt.figure(figsize=(2,10))
table = result_df[['|M1-Actual|','|M2-Actual|']]
sns.heatmap(table, center=table.mean().mean() ,cmap="Greys")
```

## Train the model we know is of good quality using the trainset

```python
X_train2 = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train2)
model.fit().summary()
```

```python
car_lm = LinearRegression()
car_lm.fit(X_train, y_train)
```

```python
# print coefficients
print('intercept ', car_lm.intercept_)
print(pd.DataFrame({'Predictor': X_train.columns, 'coefficient': car_lm.coef_}))
```

## Predict the testset using the trained model

```python
result_df['M3'] =car_lm.predict(X_test)
result_df
```

```python
result_df['|M3-Actual|'] = abs(result_df.M3-result_df.Actual)
result_df
```

```python
plt.figure(figsize=(2,10))
table = result_df[['|M1-Actual|','|M2-Actual|','|M3-Actual|']]
sns.heatmap(table, center=table.mean().mean(),cmap="Greys")
```

Now let's calculate the official metrics:

- ME: Mean Error
- RMSE: Root Mean Squared Error 
- MAE: Mean Absolute Error 
- MAPE: Mean Absolute Percentage Error

```python
metric_df = pd.DataFrame(index = ['ME','RMSE', 'MAE','MAPE'] ,columns = ['M1','M2','M3'])
metric_df
```

```python
n_test = len(result_df)

for m in metric_df.columns:
    metric_df.at['ME',m]= np.sum((result_df.Actual - result_df[m]))/n_test
    metric_df.at['RMSE',m]= np.sqrt(np.sum(result_df.apply(lambda r: (r.Actual - r[m])**2,axis=1))/n_test)
    metric_df.at['MAE',m] = np.sum(abs(result_df.Actual - result_df[m]))/n_test
    metric_df.at['MAPE',m] = np.sum(result_df.apply(lambda r:abs(r.Actual-r[m])/r.Actual,axis=1))/n_test*100
metric_df
```
